```python
import numpy as np
import pandas as pd
```

## Load the lesson streaks data

```python
lesson_stats = pd.read_csv('datasets/streak-leaderboard/LessonStreaks.csv', parse_dates=['date'])
lesson_stats
```

## Grab the user and lesson completion date, then remove duplicates

```python
user_streaks = lesson_stats[['user_id', 'user_name', 'date']].drop_duplicates()
user_streaks
```

## Group over consecutive day streaks

```python
indexes = user_streaks.groupby('user_id')['date'].rank(method='min')
date_groups = user_streaks['date'] - pd.to_timedelta(indexes, unit='D')
user_streaks['group'] = date_groups
user_streaks
```

```python
result = user_streaks.groupby(['user_id', 'user_name', 'group'], as_index=False)['date'].agg(
    start_date='min',
    end_date='max',
    streaks='count'
)
```

## List top 10 longest streaks

```python
result.nlargest(n=10, columns=['streaks'])
```

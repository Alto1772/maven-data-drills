```python
import numpy as np
import pandas as pd
```

```python
price_history = pd.read_csv('datasets/the-price-is-right/price_history.csv', parse_dates=['effective_date'])
products = pd.read_csv('datasets/the-price-is-right/products.csv')
transactions = pd.read_csv('datasets/the-price-is-right/transactions.csv', parse_dates=['order_date'])
```

```python
transactions_prices = pd.merge_asof(
    transactions.sort_values('order_date'),
    price_history.sort_values('effective_date'),
    by='pizza_id',
    left_on='order_date',
    right_on='effective_date',
    direction='backward'
)
```

```python
result = transactions_prices.merge(products, how='inner', on='pizza_id')
result = result.drop(columns=['order_detail_id', 'order_id', 'effective_date', 'current_price'])
```

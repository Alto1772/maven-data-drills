```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('datasets/splitting-the-field/baseball_positions.csv')
dataset.head()
```

```python
positions = dataset['Position'].str.split('/')
positions = positions.explode()
positions = positions.value_counts().reset_index(name='Players')
positions = positions.sort_values('Players', ascending=False)
positions
```
